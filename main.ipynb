```python
import os
import openai
from dotenv import load_dotenv
```

### Load Keys

```python
load_dotenv()
openai.api_key = os.environ['OPENAI_API_KEY']
```

```python
def generate(prompt, temperature, tokens):
    res = openai.Completion.create(
        model="text-davinci-003", 
        prompt=prompt, 
        temperature=temperature, 
        max_tokens=tokens
    )
    return res.choices[0].text[2:]

def approve(generatedText):
    print("Generated Text:")
    print(generatedText)
    print("Is this text acceptable?\n")
    return input()
```

```python
approved = False
niche = None
while approved == False:
    niche = generate(
    "randomly pick either Health and Welness, E-Commerce, or Relationships, then return back ONLY the one you have picked", 
    0.6, #temperatuzre
    20) #max tokens
    if approve(niche).lower() != 'no':
        approved = True
```

```python
title = generate(
    f"come up with a title of an ebook to captivate someone on the following niche: {niche}. be very original, creative, mention secrets",
    0.7,
    100
)
title
```

```python
ebook_outline = generate(f"generate an e-book outline for a 100 page book on the following title: {title}", 0.7, 250)
ebook_outline
```

```python
#were going to automate making each chapter of the book in a loop
chapters = []
for i in range(1, 101):
    #include the ebook outline and which chapter it is going to do
    chapter = generate(f"generate a chapter for the following ebook outline: {ebook_outline} and chapter number: {i}", 0.7, 250)
    chapters.append(chapter)

#write out the chapters to a single file with good spacing
with open("ebook.txt", "w") as f:
    for chapter in chapters:
        f.write(chapter)
        f.write("\n\n")
```
