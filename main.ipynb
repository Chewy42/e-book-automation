# Libraries

```python
import os
import openai
from dotenv import load_dotenv
```

```python
load_dotenv()
openai.api_key = os.environ['OPENAI_API_KEY']
```

```python
def generate(prompt, temperature, tokens):
    res = openai.Completion.create(
        model="text-davinci-003", 
        prompt=prompt, 
        temperature=temperature, 
        max_tokens=tokens,
        n=3
    )
    for i in res.choices:
        print(i.text[2:])
    return res.choices[0].text[2:]
```

```python
niche = generate(
    "randomly pick either Health and Welness, E-Commerce, or Relationships, then return back ONLY the one you have picked", 
    0.6, #temperature
    20) #max tokens
niche
```

```python
title = generate(
    f"come up with a title of an ebook to captivate someone on the following niche: {niche}. be very original, creative, mention secrets",
    0.7,
    50
)
title
```
